```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('movies_metadata.csv')
#Read CSV
```

```python
pd.set_option('display.max_columns', 40)
df.head()
#Display table head
```

```python
df['budget'].unique()
df['budget'].replace('.*([^0-9.]).*', '0', regex=True, inplace=True)
df['budget'].apply(float)
# finding non numerical values and setting them to 0
```

```python
df['popularity'].replace('.*([^0-9.]).*', '0', regex=True, inplace=True)
df['popularity'].apply(float)
# finding non numerical values and setting them to 0
```

```python
def fillZeroes(df, cols):
    for column in cols:
        df[column] = np.where(df[column] == 0, None, df[column]) 
    imp_mean = SimpleImputer(strategy='mean')
    df[cols] = imp_mean.fit_transform(df[cols])
#Function to impute data
```

```python
missingCols = ['budget', 'popularity', 'revenue', 'runtime', 'vote_average', 'vote_count']
fillZeroes(df, missingCols)
df[missingCols].head()
#Imputing missing data
```

```python
df['belongs_to_collection'] = df['belongs_to_collection'].apply(lambda x: str(x).split("'")[5] if len(str(x).split("'")) > 6 else None)
df.head()
#Parse Json to get collection title
```

```python
def getJsonList(js):
    try:
        js = eval(str(js))
        for j in range(len(js)):
            js[j] = js[j]['name']
        return ", ".join(js)
    except:
        return None
#Parse Json list
df['genres'] = df['genres'].apply(lambda x: getJsonList(x) if x is not None else None)
```

```python
df['production_companies'] = df['production_companies'].apply(lambda x: getJsonList(x) if x is not None else None)
df['production_countries'] = df['production_countries'].apply(lambda x: getJsonList(x) if x is not None else None)
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: getJsonList(x) if x is not None else None)
df.head()
#Validate Json parsed list
```

```python
df.to_csv('new_movies_metadata.csv')
```
